# Character Analysis

```python
# 1. Import required libraries.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# 2. Read the csv.
chars = pd.read_csv("characters_stats.csv")
```

```python
# 3. Show the first record from the csv
chars.head(1)
```

```python
# 4. Show the number of rows and columns
rows = chars.shape[0]
cols = chars.shape[1]
print(f"The chars dataframe has {rows} rows and {cols} columns.")
```

```python
# 5. Find the amount of characters who fall under each alignment, you can use value_counts()
chars["Alignment"].value_counts()
```

```python
# 6. Find only the good alignment characters
good = chars.where(chars["Alignment"] == "good").dropna().convert_dtypes()
```

```python
# 7. Show only the first five records from the previous bullet point
good.head()
```

```python
# 8. Select the top five heroes with the highest speed who have a good alignment
good_speed = good.where(good["Speed"] >= 100).dropna().count()[1]
print(f"There are {good_speed} 'good' characters with 100 Speed. These 5 have the highest Total stat.")
good.nlargest((good_speed + 1), "Speed").sort_values(by = ["Speed", "Total"], ascending = False).head()
```

```python
# 9. Select five heroes with maximum power who have a good alignment
power = good.where(good["Power"] >= 100).dropna().count()[1]
print(f"There are {power} 'good' characters with 100 Power. These 5 have the highest Total stat.")
good.nlargest((power + 1), "Power").sort_values(by = ["Power", "Total"], ascending = False).head()
```

```python
# 10.	Find out how many superheroes there are with power of 100 with good alignment
print(f"There are {power} 'good' characters with 100/100 Power.")
```

```python
# 11. Shape them what you got in point 10.
good.where(good["Power"] >= 100).dropna().shape
```

```python
# 12. Show all records from point 10.
good.nlargest(power, "Power").sort_values(by = ["Power", "Total"], ascending = False)
```

```python
# 13. Retrieve total of first five records with max power and has good alignment super heroes.
good_max_power = good.nlargest((power + 1), "Power").sort_values(by = ["Power", "Total"], ascending = False).head()
good_max_power
```

```python
# 14. Draw a bar plot of all super heroes who are having good alignment and a max power of the top five heroes only, take same object of point 13, show name and total in plot with green bars.
good_max_power_asc = good_max_power.sort_values(by="Total", ascending=True)
gmp = sns.barplot(data = good_max_power_asc[[]], x = good_max_power_asc.Name, y = good_max_power_asc.Total, color = "green")
gmp.set(title = "Highest Total Power with Good Alignment")
gmp.set_xticklabels(rotation = 25, labels = good_max_power_asc.Name)
for bar in gmp.containers:
    gmp.bar_label(bar)
gmp;
```

```python
# 15. Extract villains having bad alignment.
bad = chars.where(chars["Alignment"] == "bad").dropna()
```

```python
# 16. Show first five records of point 15.
bad.convert_dtypes().head()
```

```python
# 17. Show top five fastest super villains in terms of super speed.
bad_speed = bad.where(bad["Speed"] >= 100).dropna().count()[1]
print(f"There are {bad_speed} 'bad' characters with 100 Speed, so Top 4 are organized by Total.")
bad.nlargest(5, "Speed").sort_values(by=["Speed", "Total"], ascending=False).convert_dtypes().head()
```

```python
# 18. Top five super villains in terms of intelligence.
intel = bad.where(bad["Intelligence"] >= 100).dropna().count()[1]
print(f"There are {intel} 'bad' characters with 100 Intelligence (or higher). Spots 2-4 are organized by Total.")
bad.nlargest((intel + 1), "Intelligence").sort_values(by = ["Intelligence", "Total"], ascending = False).convert_dtypes().head()
```

```python
# 19. Show who is most dangerous super villain after calculating their total (top 5 only).
bad.nlargest(5, "Total").sort_values(by=["Total"], ascending=False).convert_dtypes()
```

```python
# 20. Draw a histogram for speed of super heroes having fig size 10,5, provide speed in histogram for only good alignment super heroes, title should be "distribution of speed", xlabel should be "speed".
good_max_speed = good.nlargest(10, "Speed").sort_values(by = ["Speed", "Total"], ascending = True)

gms = sns.barplot(data=good_max_speed, x = good_max_speed.Name, y = good_max_speed.Speed)
gms.set(title = "Distribution of Speed")
gms.set_xticklabels(rotation = 30, labels = good_max_speed.Name)
for bar in gms.containers:
    gms.bar_label(bar)
gms;
```

```python
# 21. Draw a histogram for combat of super villains having fig size 10,5, provide combat in histogram for only bad alignment super heroes, title should be "distribution of combat", xlabel should be "combat".
combat = bad.where(bad["Combat"] >= 100).dropna()
print(f"There are {combat.count()[1]} 'bad' characters with 100 Combat.")
bad_max_combat = bad.nlargest(10, "Combat").sort_values(by = ["Combat", "Total"], ascending=True)

bmc = sns.barplot(data = bad_max_combat, x = bad_max_combat.Name, y = bad_max_combat.Combat)
bmc.set(title = "Distribution of Combat")
bmc.set_xticklabels(rotation = 30, labels = bad_max_combat.Name)
for bar in bmc.containers:
    bmc.bar_label(bar)
bmc;
```

```python
# I made a list of characters who are "overpowered" (have a stat above 100)

# Gives characters tiers based on "Total." Gives a list of conditions to check and a list of values to assign using np.select().
conditions = [
    (chars["Total"] >= 567),
    (chars["Total"] >= 534),
    (chars["Total"] >= 500),
    (chars["Total"] >= 467),
    (chars["Total"] >= 434),
    (chars["Total"] >= 400),
    (chars["Total"] >= 367),
    (chars["Total"] >= 334),
    (chars["Total"] >= 300),
    (chars["Total"] >= 267),
    (chars["Total"] >= 234),
    (chars["Total"] >= 200),
    (chars["Total"] >= 167),
    (chars["Total"] >= 134),
    (chars["Total"] >= 100),
    (chars["Total"] >= 67),
    (chars["Total"] >= 34),
    (chars["Total"] >= 0),
    ]

values = ['S+','S','S-','A+','A','A-','B+','B','B-','C+','C','C-','D+','D','D-','F+','F','F-']

chars['Tier'] = np.select(conditions, values)
chars
```

```python
# Shows the distribution of different tiers in the chars dataset
tierDist = sns.countplot(x = chars["Tier"], order = ['F-','F','F+','D-','D','D+','C-','C','C+','B-','B','B+','A-','A','A+','S-','S','S+'])
tierDist.set(title = "Tier Distribution")
for bar in tierDist.containers:
    tierDist.bar_label(bar)
```

```python
# Same as above without all the F- Tier characters (they seem to mostly be outliers/bad data)
tierDist = sns.countplot(x = chars["Tier"], order = ['F','F+','D-','D','D+','C-','C','C+','B-','B','B+','A-','A','A+','S-','S','S+'])
tierDist.set(title = "Tier Distribution")
for bar in tierDist.containers:
    tierDist.bar_label(bar)
```

```python
# Function to return characters of a specific tier
def charByTier(tier):
  if tier.upper() in values: 
    return chars.loc[chars['Tier'] == tier.upper()].sort_values(by = "Total", ascending = False)
  else: 
    print(f"Tier must be one of the following from the list: {values}.")

charByTier("D-")
```
