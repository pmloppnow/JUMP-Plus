```python
# 2. Import any required libraries.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 3. Read the cleaned csv from phase 1.
sales = pd.read_csv("Video_Games_Sales_Cleaned.csv")
```

```python
# 7. Show the last 5 five rows.
sales.tail()
```

```python
# 8. Describe the data.
sales.describe().round(2)
```

```python
# 9. Group the sales data of each region on Year_of_Release and plot the same (It should show which regions sales were at peak on which year of release)
year = sales.groupby("Year_of_Release")[["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales"]].sum()
year_sales = year.plot(kind = "bar", rot = 90)
plt.title("Sales by Year of Release")
plt.xlabel("Year of Release")
plt.ylabel("Sales")
year_sales.legend(loc = "upper left")
plt.show()
```

```python
# 10. Group North America, Japan, Europe & Other regions sales data by Genre and plot it (It should show highest of region according to game category, 
# ex. NA_Sales is highest in Adventure category, use a barplot)
genre = sales.groupby("Genre")[["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales"]].sum()
genre_sales = genre.plot(kind = "bar", rot = 45)
plt.title("Sales by Genre")
plt.xlabel("Genre")
plt.ylabel("Sales (Millions)")
genre_sales.legend(loc = "upper center")
plt.show()
```

```python
# 11. Group North America, Japan, Europe & other regions sales data per Platform and plot it (Should show sales as per "Platform genre", use a barplot)
platform = sales.groupby("Platform")[["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales"]].sum()
platform_sales = platform.plot(kind = "bar", rot = 90)
plt.title("Sales by Platform")
plt.xlabel("Platform")
plt.ylabel("Sales (Millions)")
platform_sales.legend(loc = "upper left")
plt.show()
```

```python
# 12. Plot the Sales in all regions per Platform, you need to create a boxplot, which will show the distribution of sales per platform.
sns.boxplot(data = platform)
plt.title("Sales per Platform")
plt.xlabel("Country")
plt.ylabel("Sales (Millions)")
plt.show()
```

```python
# 13. Find out which Genre of games are sold in highest numbers. (Create a countplot using seaborn)
highest_genre = sns.countplot(x = "Genre", data = sales)
plt.title("Game Count by Genre")
plt.xlabel("Genre")
plt.ylabel("Distinct Game Count")
plt.xticks(rotation = 45)
for bar in highest_genre.containers:
  highest_genre.bar_label(bar)
plt.show()
```

```python
# 14. Try to find out global sales per Genre (Create a barplot using seaborn)
global_sales = sns.barplot(x = "Genre", y = "Global_Sales", data = sales, estimator = "sum", errorbar = None) 
plt.title("Global Sales by Genre")
plt.ylabel("Global Sales (Millions)")
plt.xticks(rotation = 45)
for bar in global_sales.containers:
  global_sales.bar_label(bar)
plt.show;
```
